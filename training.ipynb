```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.layers import Dense,SimpleRNN,Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
reviews = [
    "movie is very good",
    "film was excellent",
    "movie is bad",
    "film was terrible"
]

# 1 = Positive, 0 = Negative
labels = [1, 1, 0, 0]
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(reviews)#it learn the vocabulary from the text and give the each word a number ID
sequence = tokenizer.texts_to_sequences(reviews)#It replaces words with their numbers
```

```python
X = pad_sequences(sequence,maxlen=4)
y = np.array(labels)

word_size = len(tokenizer.word_index) + 1 #It is a dictionary of all words learned by tokenizer. (+1 Index 0 is reserved for padding so we add +1,This value is used in the Embedding layer)
```

RNN

```python
model = Sequential([
    Embedding(input_dim = word_size,output_dim = 8,input_length = 4),
    SimpleRNN(6),
    Dense(1,activation = 'sigmoid')
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.2),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=10,verbose=1)
```

```python
#prediction for new data

test_review = input("enter the movie review:")
test_seq = tokenizer.texts_to_sequences([test_review])
test_pad = pad_sequences(test_seq, maxlen=4)

prediction = model.predict(test_pad)
score = prediction[0][0]

print("Sentiment:","Positive" if score > 0.5 else "Negative")
```
